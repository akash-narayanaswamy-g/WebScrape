```python
import requests
from bs4 import BeautifulSoup
import csv

# List of Amazon product page URLs
urls = [
    "https://www.amazon.in/dp/<ASIN_1>",
    "https://www.amazon.in/dp/<ASIN_2>",
    "https://www.amazon.in/dp/<ASIN_3>",
    # add more URLs here
]

# Open the CSV file for writing
with open("product_data.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Brand", "Price", "Rating", "Reviews"])

    # Iterate over each URL and extract the product information
    for url in urls:
        try:
            # Send a GET request to the product page URL
            response = requests.get(url)

            # Parse the HTML content using Beautiful Soup
            soup = BeautifulSoup(response.content, "html.parser")

            # Extract the product title
            title_elem = soup.find("span", {"id": "productTitle"})
            title = title_elem.text.strip() if title_elem else ""

            # Extract the product brand
            brand_elem = soup.find("a", {"id": "bylineInfo"})
            brand = brand_elem.text.strip() if brand_elem else ""

            # Extract the product price
            price_elem = soup.find("span", {"id": "priceblock_ourprice"})
            price = price_elem.text.strip() if price_elem else ""

            # Extract the product rating
            rating_elem = soup.find("span", {"class": "a-icon-alt"})
            rating = rating_elem.text.strip() if rating_elem else ""

            # Extract the product review count
            reviews_elem = soup.find("span", {"id": "acrCustomerReviewText"})
            reviews = reviews_elem.text.strip() if reviews_elem else ""

            # Write the data to the CSV file
            writer.writerow([title, brand, price, rating, reviews])

            # Print a success message
            print(f"Successfully scraped data for URL: {url}")

        except Exception as e:
            # Handle any errors that occur during scraping
            print(f"Error scraping data for URL: {url}")
            print(str(e))
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd


columns = ['Description', 'ASIN', 'Product Description', 'Manufacturer']

# Define the base URL for the product listings page
base_url = 'https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1'

# Use requests to retrieve the HTML content of the page
response = requests.get(base_url)
html_content = response.content

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all the product blocks on the page
product_blocks = soup.find_all('div', {'data-component-type': 's-search-result'})

# Create a list to store the data for each product
products_data = []
df = pd.DataFrame(columns=columns)

# Iterate over each product block and extract the product name, price and URL
for product_block in product_blocks:
    product_name = product_block.find('span', {'class': 'a-text-normal'}).text.strip()
    product_price = product_block.find('span', {'class': 'a-price-whole'}).text.strip()
    product_url = 'https://www.amazon.in' + product_block.find('a', {'class': 'a-link-normal'})['href']
    
    # Use requests to retrieve the HTML content of the product page
    product_response = requests.get(product_url)
    product_html_content = product_response.content
    
    # Parse the HTML content using Beautiful Soup
    product_soup = BeautifulSoup(product_html_content, 'html.parser')
    
    # Extract the ASIN, product description, manufacturer and other required information
    asin = product_soup.find('span', {'class': 'a-size-base', 'id': 'ASIN'}).text.strip()
    product_description = product_soup.find('div', {'id': 'productDescription'}).text.strip()
    manufacturer = product_soup.find('a', {'id': 'bylineInfo'}).text.strip()
    product_details = product_soup.find('div', {'id': 'detailBullets_feature_div'}).find_all('li')
    description = ''
    for product_detail in product_details:
        if 'Item Weight' in product_detail.text:
            weight = product_detail.text.strip()
            description += f'{weight}, '
        elif 'Product Dimensions' in product_detail.text:
            dimensions = product_detail.text.strip()
            description += f'{dimensions}, '
    
    # Add the extracted data to the products_data list
    products_data.append({
        'Product Name': product_name,
        'Product Price': product_price,
        'Product URL': product_url,
        'ASIN': asin,
        'Product Description': product_description,
        'Manufacturer': manufacturer,
        'Description': description
    })
    
    data = {'Description': description, 'ASIN': asin, 'Product Description': product_description, 'Manufacturer': manufacturer}

    # Add the dictionary to the DataFrame
    df = df.append(data, ignore_index=True)
    
    print(products_data)
    # Print the extracted data for each product
    print(f"Product Name: {product_name}")
    print(f"Product Price: {product_price}")
    print(f"Product URL: {product_url}")
    print(f"ASIN: {asin}")
    print(f"Product Description: {product_description}")
    print(f"Manufacturer: {manufacturer}")
    print(f"Description: {description}")
    print()
```
