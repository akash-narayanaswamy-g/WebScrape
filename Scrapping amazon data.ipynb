```python
import time
import csv
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

chrome_path = r"/Library/Frameworks/Python.framework/Versions/3.6/bin/chromedriver"
driver = webdriver.Chrome(chrome_path)
# Set the search query and the base URL
query = 'bags'
base_url = f'https://www.amazon.in/s?k={query}'

# Set the number of products to scrape
num_products = 200
try:
    # Open the CSV file for writing
    with open('amazon_products.csv', mode='w', newline='', encoding='utf-8') as file:

        # Create the CSV writer
        writer = csv.writer(file)

        # Write the header row
        writer.writerow(['Description', 'ASIN', 'Product Description', 'Manufacturer'])

        # Set up the web driver
        options = webdriver.ChromeOptions()
        options.add_argument('--headless')  # Uncomment this line to run the script in headless mode
        driver = webdriver.Chrome(options=options)

        # Load the first page of search results
        driver.get(base_url)

        # Keep track of the number of products scraped so far
        num_scraped = 0

        # Loop through each page of search results
        while num_scraped < num_products:

            # Scroll to the bottom of the page to load more products
            driver.find_element_by_tag_name('body').send_keys(Keys.END)
            time.sleep(2)

            # Check if the "Next" button is available
            try:
                next_button = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.LINK_TEXT, 'Next'))
                )
                next_url = next_button.get_attribute('href')
            except NoSuchElementException:
                # If there is no "Next" button, we have reached the end of the search results
                break

            # Load the next page of search results
            driver.get(next_url)

            # Find all the product cards on the page
            product_cards = driver.find_elements_by_css_selector('div[data-asin]')

            # Loop through each product card and extract the required information
            for card in product_cards:
                try:
                    # Get the product title and ASIN
                    title_element = card.find_element_by_css_selector('h2 a')
                    title = title_element.get_attribute('title')
                    asin = card.get_attribute('data-asin')

                    # Get the product description and manufacturer
                    description_element = card.find_element_by_css_selector('div.a-section.a-spacing-small')
                    description_text = description_element.text
                    manufacturer_element = description_element.find_elements_by_css_selector('span')[1]
                    manufacturer = manufacturer_element.text

                    # Write the data to the CSV file
                    writer.writerow([title, asin, description_text, manufacturer])

                    # Increment the counter
                    num_scraped += 1

                    # Check if we have scraped enough products
                    if num_scraped >= num_products:
                        break

                except NoSuchElementException:
                    # If any of the required elements are not found, skip this product
                    continue

        
except Exception as e:
    print("Driver running !!!")
    # Close the web driver
    driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Set the search query and the base URL
query = 'bags'
base_url = f'https://www.amazon.in/s?k={query}'

# Set the number of products to scrape
num_products = 200

# Open the CSV file for writing
with open('amazon_products.csv', mode='w', newline='', encoding='utf-8') as file:

    # Create the CSV writer
    writer = csv.writer(file)

    # Write the header row
    writer.writerow(['Description', 'ASIN', 'Product Description', 'Manufacturer'])

    # Keep track of the number of products scraped so far
    num_scraped = 0

    # Loop through each page of search results
    while num_scraped < num_products:

        # Load the search results page
        response = requests.get(base_url)

        # Parse the HTML response using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all the product cards on the page
        product_cards = soup.select('div[data-asin]')

        # Loop through each product card and extract the required information
        for card in product_cards:
            try:
                # Get the product title and ASIN
                title_element = card.select_one('h2 a')
                title = title_element.get('title')
                asin = card.get('data-asin')

                # Get the product description and manufacturer
                description_element = card.select_one('div.a-section.a-spacing-small')
                description_text = description_element.get_text(strip=True)
                manufacturer_element = description_element.select('span')[1]
                manufacturer = manufacturer_element.get_text(strip=True)

                # Write the data to the CSV file
                writer.writerow([title, asin, description_text, manufacturer])

                # Increment the counter
                num_scraped += 1

                # Check if we have scraped enough products
                if num_scraped >= num_products:
                    break

            except (AttributeError, IndexError):
                # If any of the required elements are not found, skip this product
                continue

        # Find the "Next" button URL
        next_button = soup.select_one('a#pagnNextLink')
        if next_button:
            next_url = next_button.get('href')
        else:
            # If there is no "Next" button, we have reached the end of the search results
            break

        # Set the base URL for the next page of search results
        base_url = f'https://www.amazon.in{next_url}'

    # Close the CSV file
    file.close()
```
